```python
# Get the hisotrical weanther data for games played
import pandas as pd
import requests
import time
import datetime
from pprint import pprint
from mike_api import api_key
# create code to test the WorldWeatherOnline API


# load and read csv data into dataframe
game_data_toload = "stadiums_merge_type.csv"
game_data = pd.read_csv(game_data_toload)
game_data.head()

# these two series will hold the data to create API call's
cityid = game_data['CITY']
game_date =[]

# change dates to propoer format for the API request URL
for game in game_data['schedule_date']:
    game_date.append(datetime.datetime.strptime(game, '%m/%d/%Y').strftime('%Y-%m-%d'))
    
#game_date
```

```python
# Base URL for the worldwide weather onine API
url = 'https://api.worldweatheronline.com/premium/v1/past-weather.ashx?q='

# arrays for holding the data retreived
temp = []
description = []
weather_code = []
location = []
date = []
count = 0

#loop through the cities and dates from imported csv file/dataframe to request weather data for selected games
for city in cityid:
    try:
        #print(game_date[count])
        d = game_date[count] 
        request_url = f"{url}{city},US&date={d}&format=json&key={api_key}"
        count += 1
        weather = requests.get(request_url).json()
        location.append(city)
        date.append(d)
        temp.append(weather['data']['weather'][0]['hourly'][6]['FeelsLikeF'])
        description.append(weather['data']['weather'][0]['hourly'][6]['weatherDesc'][0]['value'])
        weather_code.append(weather['data']['weather'][0]['hourly'][6]['weatherCode'])
        print(request_url)
    except:
        print("No game in City this week")
        
print(temp)
print(description)
print(weather_code)
```

```python
# Create a dictionary from the weather data lists. Used to create dataframe


weather_dict = {
    'Game Date': date,
    'City': location,
    'Temperature': temp,
    'Weather': description, 
    'Code': weather_code
}
#weather_dict

# create a data frame from arrays filled by responses
weather_df = pd.DataFrame(weather_dict)
weather_df.to_csv('weather_data.csv')
weather_df
```
